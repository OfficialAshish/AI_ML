```python
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import fetch_california_housing
data_ = fetch_california_housing(as_frame=True)
# print(data_.DESCR)
data_.data.head()
```

```python
# data = pd.DataFrame(data_["data"], columns = data_["feature_names"])
# data["target"] = pd.Series( data_["target"])
# data.head()
X = data_.data
y = data_.target
y.head()
```

```python
X.shape, y.shape;
```

```python
# getting data ready

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
X_train, y_train;
X_test, y_test;
```

```python
# Lets do the main part here
from sklearn.linear_model import Ridge
model = Ridge()
model.fit(X_train, y_train)
score = model.score(X_test, y_test)
print("Model Score:", score)

para = model.coef_, model.intercept_
print("Feature Parameters:", para)
```

```python
import matplotlib.pyplot as plt

# Get feature names (assuming X_train is a DataFrame or array with named columns)
feature_names = X_train.columns

# Get coefficients and intercept from the model
coefficients = model.coef_
intercept = model.intercept_

# Plot coefficients
plt.figure(figsize=(10, 6))
plt.barh(feature_names, coefficients)
plt.xlabel('Coefficient Value')
plt.ylabel('Feature')
plt.title('Ridge Regression Coefficients')
plt.axvline(x=0, color='gray', linestyle='--', linewidth=0.5)  # Add vertical line at 0
plt.show();
```

```python
 
import matplotlib.pyplot as plt

# Extract the feature for the x-axis
x_feature = 'MedInc'

# Plot actual data points
plt.scatter(X_test[x_feature], y_test, color='blue', label='Actual Data')

predicted_values = model.predict(X_test)

# Plot predicted values based on the linear regression model
plt.scatter(X_test[x_feature], predicted_values, color='red', label='Predicted Data')

# Plot the linear regression line
plt.plot(X_test[x_feature], predicted_values, color='green', label='Regression Line')

# Add labels and legend
plt.xlabel(x_feature)
plt.ylabel('Target Variable')
plt.title('Scatter Plot with Linear Regression Prediction Line')
plt.legend()

# Show plot
plt.show()
```

```python
# Choosing another model 
from sklearn.ensemble import RandomForestRegressor
np.random.seed(3)
rf = RandomForestRegressor()
rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```

```python
 
# Extract the feature for the x-axis
x_feature = 'MedInc'

fig , ax  = plt.subplots(figsize=(8,5))


# Plot actual data points
ax.scatter(X_test[x_feature], y_test, color='blue', label='Actual Data')

predicted_values = rf.predict(X_test)

# Plot predicted values based on the linear regression model
ax.scatter(X_test[x_feature], predicted_values, color='red', label='Predicted Data')

# Plot the linear regression line
ax.plot(X_test[x_feature], predicted_values, color='green', label='Regression Line')

# Add labels and legend
ax.set(xlabel =x_feature,
    ylabel = 'Target Variable',
    title='Scatter Plot with Linear Regression Prediction Line'
      )
ax.legend()
```
