
# Using keras dataset of digits classification

```python
import tensorflow as tf
```

```python
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.mnist.load_data()
```

```python
X_train.shape,y_train.shape
```

```python
X_train
```

```python
import matplotlib.pyplot as plt
plt.imshow(X_train[24])
```

```python
X_train = X_train/255
X_test = X_test/255
```

```python
model = tf.keras.Sequential()

model.add(tf.keras.layers.Flatten(input_shape=(28,28)))
model.add(tf.keras.layers.Dense(128,activation='relu'))
model.add(tf.keras.layers.Dense(32,activation='relu'))
model.add(tf.keras.layers.Dense(10,activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='sparse_categorical_crossentropy',optimizer='Adam',metrics=['accuracy'])
```

```python
history = model.fit(X_train,y_train,epochs=10,validation_split=0.2)
```

```python
y_preds = model.predict(X_test)
```

```python
y_preds
```

```python
y_preds = y_preds.argmax(axis=1)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_preds)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
```

```python
import numpy as np
import matplotlib.pyplot as plt
def plot_prediction(y_preds = y_preds):
  n_test_images = X_test.shape[0]
  random_indices = np.random.randint(0, n_test_images, size=9)

  # Get the predictions and true labels for the selected images
  predictions = y_preds[random_indices]
  true_labels = y_test[random_indices]

  fig, axes = plt.subplots(3, 3, figsize=(10, 10))

  for i, ax in enumerate(axes.flat):

    image = X_test[random_indices[i]].reshape(28, 28)

    # Display the image
    ax.imshow(image, cmap="gray")

    ax.set_title(f"Predicted: {predictions[i]}, True: {true_labels[i]}", fontsize=10)
    ax.axis('off')


  plt.show()
```

```python
plot_prediction()
```

```python
# Trying another nn model
model_1 = tf.keras.Sequential()

model_1.add(tf.keras.layers.Flatten(input_shape=(28,28)))
model_1.add(tf.keras.layers.Dense(128,activation='relu'))
model_1.add(tf.keras.layers.Dropout(0.2))
model_1.add(tf.keras.layers.Dense(64,activation='relu'))
model_1.add(tf.keras.layers.Dropout(0.2))
model_1.add(tf.keras.layers.Dense(32,activation='relu'))
model_1.add(tf.keras.layers.Dropout(0.2))
model_1.add(tf.keras.layers.Dense(10,activation='softmax'))
model_1.summary()
model_1.compile(loss='sparse_categorical_crossentropy',optimizer='Adam',metrics=['accuracy'])
history_1 = model_1.fit(X_train,y_train,epochs=10,validation_split=0.2)
```

```python
y_preds_1 = model_1.predict(X_test)
y_preds_1 = y_preds_1.argmax(axis=1)
accuracy_score_1 = accuracy_score(y_test, y_preds_1)
print(f"Accuracy of model_1: {accuracy_score_1}")
```

```python
plot_prediction(y_preds_1)
```

```python

def plot_incorrect_predictions(y_preds=y_preds, y_test=y_test):
  incorrect_indices = np.nonzero(y_preds != y_test)[0]
  n_incorrect_images = len(incorrect_indices)

  # Select a random sample of incorrect predictions
  random_indices = np.random.randint(0, n_incorrect_images, size=9)

  predictions = y_preds[incorrect_indices[random_indices]]
  true_labels = y_test[incorrect_indices[random_indices]]

  fig, axes = plt.subplots(3, 3, figsize=(10, 10))

  for i, ax in enumerate(axes.flat):
    image = X_test[incorrect_indices[random_indices[i]]].reshape(28, 28)

    # Display the image
    ax.imshow(image, cmap="gray")

    ax.set_title(f"Predicted: {predictions[i]}, True: {true_labels[i]}", fontsize=10)
    ax.axis('off')

  plt.show()
```

```python

plot_incorrect_predictions(y_preds_1)
```
